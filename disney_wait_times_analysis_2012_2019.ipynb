```python
def load_and_clean_ride(file_name, ride_name):
    """
    Loads and cleans a Disney ride wait time dataset.

    Parameters:
    file_name (str): Name of the CSV file
    ride_name (str): Name of the ride

    Returns:
    DataFrame: Cleaned dataset with standardized columns
    """

    # Load the dataset
    df = pd.read_csv(file_name)

    # Convert 'datetime' column to proper datetime format
    df['datetime'] = pd.to_datetime(df['datetime'])

    # Drop rows where BOTH posted and actual wait times are missing
    df = df.dropna(subset=['SPOSTMIN', 'SACTMIN'], how='all')

    # Create a unified wait time column
    # Prioritize actual wait time, fallback to posted wait time
    df['wait_time'] = df['SACTMIN'].fillna(df['SPOSTMIN'])

    # Add ride name column
    df['ride_name'] = ride_name

    return df
```

```python
splash = load_and_clean_ride("splash_mountain.csv", "Splash Mountain")
pirates = load_and_clean_ride("pirates_of_caribbean.csv", "Pirates of the Caribbean")
everest = load_and_clean_ride("expedition_everest.csv", "Expedition Everest")
safaris = load_and_clean_ride("kilimanjaro_safaris.csv", "Kilimanjaro Safaris")
navi = load_and_clean_ride("navi_river.csv", "Na'vi River Journey")
dwarfs = load_and_clean_ride("dwarfs_train.csv", "Seven Dwarfs Mine Train")
```

```python
df_all = pd.concat([splash, pirates, everest, safaris, navi, dwarfs], ignore_index=True)

df_all.head()
```

```python
# Extract day of week (Monday, Tuesday, etc.)
df_all['day_of_week'] = df_all['datetime'].dt.day_name()

# Extract hour of day (0–23)
df_all['hour'] = df_all['datetime'].dt.hour

df_all.head()
```

```python
def get_time_of_day(hour):
    if hour < 12:
        return "Morning"
    elif hour < 17:
        return "Afternoon"
    else:
        return "Evening"

df_all['time_of_day'] = df_all['hour'].apply(get_time_of_day)
```

```python
avg_wait_by_ride = df_all.groupby('ride_name')['wait_time'].mean().sort_values(ascending=False)

avg_wait_by_ride
```

```python
# Remove negative values
df_all = df_all[df_all['wait_time'] >= 0]

# Remove missing values
df_all = df_all.dropna(subset=['wait_time'])

# Double check
df_all['wait_time'].min()
```

```python
# Recreate time features AFTER cleaning
df_all['day_of_week'] = df_all['datetime'].dt.day_name()
df_all['hour'] = df_all['datetime'].dt.hour

def get_time_of_day(hour):
    if hour < 12:
        return "Morning"
    elif hour < 17:
        return "Afternoon"
    else:
        return "Evening"

df_all['time_of_day'] = df_all['hour'].apply(get_time_of_day)
```

```python
wait_by_time = df_all.groupby('time_of_day')['wait_time'].mean()
wait_by_time
```

```python
avg_wait_by_ride = df_all.groupby('ride_name')['wait_time'].mean()
avg_wait_by_ride
```

```python
df_all['lightning_lane_wait'] = df_all['wait_time'] * 0.3
```

```python
comparison = df_all[['wait_time', 'lightning_lane_wait']].mean()
comparison
```

```python
ll_by_ride = df_all.groupby('ride_name')[['wait_time', 'lightning_lane_wait']].mean()
ll_by_ride
```

```python
avg_wait_by_ride = avg_wait_by_ride.sort_values(ascending=False)

avg_wait_by_ride.plot(
    kind='bar',
    figsize=(7,4),
    color='#4E79A7'
)

plt.title("Seven Dwarfs and Na'vi River Have the Longest Wait Times")
plt.ylabel("Minutes")
plt.xticks(rotation=30)
plt.ylim(0, 100)

for i, v in enumerate(avg_wait_by_ride):
    plt.text(i, v + 1, f"{v:.0f}", ha='center')

plt.savefig("avg_wait_by_ride.png", bbox_inches='tight')

plt.show()
```
